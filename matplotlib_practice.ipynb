```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set x-axis to list of strings for month
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set y-axis to list of floats as total fare in US dollars for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 adn 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add th elegend
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 adn 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston')
# Create labels for the axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 adn 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
plt.subplots(figsize=(8, 8))
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
